#**Ciclos**

```python
#if
temperatura = 30

if temperatura > 25:
    print("Hace calor")
```

```python
#else
edad = 21

if edad >= 18:
    print("Edad válida")
else:
    print("Debes ser mayor de 18")
```

```python
#elif
nota = 72

if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Aprobado")
else:
    print("Reprobado")
```

# **Condiciones**

```python
# while
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1
```

```python
#for
for letra in "Hola":
    print(letra)
```
